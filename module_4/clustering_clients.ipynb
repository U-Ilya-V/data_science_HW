### Описание признаков
##### Демографические данные

    CustomerID: Уникальный идентификатор клиента.
    BirthYear: Год рождения клиента.
    EducationLevel: Уровень образования клиента.

##### Поведение

    DaysSinceLastPurchase: Количество дней с момента последней покупки клиента.
    FiledComplaint: Указано, подавал ли клиент жалобы за последние 2 года (1 — да, 0 — нет).

##### Траты на продукты

    SpentOnWines: Сумма, потраченная на покупку вина за последние 2 года.
    SpentOnFruits: Сумма, потраченная на покупку фруктов за последние 2 года.

    SpentOnGold: Сумма, потраченная на покупку золота за последние 2 года.

##### Маркетинговые кампании

    DiscountedPurchases: Количество покупок со скидкой.
    AcceptedCampaign1: Указано, принял ли клиент предложение первой кампании (1 — да, 0 — нет).
    AcceptedCampaign2: Указано, принял ли клиент предложение второй кампании (1 — да, 0 — нет).
    AcceptedCampaign3: Указано, принял ли клиент предложение третьей кампании (1 — да, 0 — нет).
    AcceptedCampaign4: Указано, принял ли клиент предложение четвертой кампании (1 — да, 0 — нет).
    AcceptedCampaign5: Указано, принял ли клиент предложение пятой кампании (1 — да, 0 — нет).
    LastCampaignResponse: Указано, принял ли клиент предложение последней кампании (1 — да, 0 — нет).

##### Каналы покупок

    OnlinePurchases: Количество покупок, совершенных через веб-сайт компании.
    CatalogPurchases: Количество покупок, совершенных через каталог.
    InStorePurchases: Количество покупок, совершенных в магазине.
    WebsiteVisitsLastMonth: Количество визитов на веб-сайт компании за последний месяц.

##### Финансовые метрики

    ContactCost: Стоимость контакта с клиентом (постоянная метрика).
    RevenueGenerated: Доход, сгенерированный клиентом (постоянная метрика).

```python
def outlier_to_median(data,column,threshold=3):     
    IQR = data[column].quantile(0.75) - data[column].quantile(0.25)
    low_border = data[column].quantile(0.25) - (IQR * threshold)
    upper_border = data[column].quantile(0.75) + (IQR * threshold)
    outliers = pd.concat([data[column]>upper_border,data[column]<low_border],axis=1).any(axis=1)
    outliers_index = data[outliers].index
    data_copy = data.copy(deep=True)
    if len(outliers_index) > 0:
        data_copy.loc[outliers_index,column] = data_copy[column].median()
    return data_copy
```

Объеденим столбцы AcceptedCampaign и заменим их числами от 0 до 6

```python
def aggAcceptCampaing(row):
    key_value_map = {
        'AcceptedCampaign1': 6,
        'AcceptedCampaign2': 5,
        'AcceptedCampaign3': 4,
        'AcceptedCampaign4': 3,
        'AcceptedCampaign5': 2,
        'LastCampaignResponse': 1
    }
    
    for key, value in key_value_map.items():
        if row.get(key) == 1:
            return value
    return 0
```

```python
def elbow_and_dend_graphics(scaled_df):
    distance_mat = pdist(scaled_df) 

    Z = hierarchy.linkage(distance_mat, 'ward') 
    plt.figure(figsize=(6, 5))
    dn = hierarchy.dendrogram(Z, leaf_font_size=8., leaf_rotation=90.)
    plt.show()

    plt.figure(figsize=(6, 5))
    last = Z[-10:, 2]
    last_rev = last[::-1]
    idxs = np.arange(1, len(last) + 1)
    plt.plot(idxs, last_rev)

    acceleration = np.diff(last, 2)
    acceleration_rev = acceleration[::-1]
    plt.plot(idxs[:-2] + 1, acceleration_rev) 
    plt.show()

    plt.figure(figsize=(6, 5))
    score = []
    k = range(2, 10)

    for k_i in k:
        km = KMeans(n_clusters=k_i).fit(scaled_df)
        km.fit(scaled_df)
        score.append(silhouette_score(scaled_df, km.labels_))
        
    plt.plot(k, score)
    plt.show()
    
```

### KMeans

```python
from sklearn.cluster import KMeans

n_clusters_1 = 5
km_1 = KMeans(n_clusters=n_clusters_1,
            init='k-means++',
            n_init=10,
            max_iter=300,
            tol=1e-04,
            random_state=0,
            verbose=1);

y_km_1 = km_1.fit_predict(tsne_df_1)

n_clusters_2=4
km_2 = KMeans(n_clusters=n_clusters_2,
            init='k-means++',
            n_init=10,
            max_iter=300,
            tol=1e-04,
            random_state=0,
            verbose=1);

y_km_2 = km_2.fit_predict(tsne_df_2)
```

### MiniBatchKMeans

```python
from sklearn.cluster import MiniBatchKMeans
mbkm_1 = MiniBatchKMeans(n_clusters= n_clusters_1)
y_mbkm_1 = mbkm_1.fit_predict(tsne_df_1)
mbkm_2 = MiniBatchKMeans(n_clusters= n_clusters_2)
y_mbkm_2 = mbkm_2.fit_predict(tsne_df_2)
```

### AffinityPropagation

```python
from sklearn.cluster import AffinityPropagation

ap = AffinityPropagation(random_state=5)
y_ap_1 = ap.fit_predict(tsne_df_1)
y_ap_2 = ap.fit_predict(tsne_df_2)
```
